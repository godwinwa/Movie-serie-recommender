```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn import neighbors
```

```python
data = pd.read_csv('netflix_titles.csv')
```

```python
data.head()
```

```python
genres = pd.get_dummies(data["listed_in"])
```

```python
data_ext = data[["title", "release_year"]]
```

```python
data_ext.head()
```

```python
train_data = pd.concat([data_ext, genres], axis = 1)

train_data.head()
```

```python
train_data.set_index("title", inplace = True)
```

```python
train_data.head()
```

```python
scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(train_data)
```

```python
model = neighbors.NearestNeighbors(n_neighbors = 6)

model.fit(scaled_data)

dist, idlist = model.kneighbors(scaled_data)
```

```python
def recommending_per_watch_history(name):
    netflix_list = []
    netflix = data[data["title"] == name].index

    netflix = netflix[0]
    for id in idlist[netflix]:
        netflix_list.append(data.loc[id].title)
    if name in netflix_list:
        netflix_list.remove(name)
    return netflix_list
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
movies = df[(df['type'] == "Movie")]
tv_show = df[(df['type'] == "TV Show")]
```

```python
movies['duration'] = movies['duration'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)").astype(float)
tv_show['duration'] = tv_show['duration'].str.extract("([-+]?\d*\.\d+|[-+]?\d+)").astype(float)
```

```python
movies_df = df[[ 'title', 'director', 'cast', 'description', 'country', 'rating','listed_in']]
tv_show_df = df[[ 'title', 'director', 'cast', 'description', 'country', 'rating','listed_in']]
```

```python
movies_df['director'] = movies_df['director'].fillna('')
movies_df['cast'] = movies_df['cast'].fillna('')
movies_df['country'] = movies_df['country'].fillna('')
movies_df['rating'] = movies_df['rating'].fillna('')

tv_show_df['director'] = movies_df['director'].fillna('')
tv_show_df['cast'] = movies_df['cast'].fillna('')
tv_show_df['country'] = movies_df['country'].fillna('')
tv_show_df['rating'] = movies_df['rating'].fillna('')
```

```python
movies_df['cast'] = movies_df['cast'].apply(lambda x:x.split(", ")[0:10])
tv_show_df['cast'] = tv_show_df['cast'].apply(lambda x:x.split(", ")[0:10])
```

```python
movies_df['description'] = movies_df['description'].apply(lambda x:x.split())
movies_df['rating'] = movies_df['rating'].apply(lambda x:x.split())
movies_df['country'] = movies_df['country'].apply(lambda x:x.split(","))
movies_df['director'] = movies_df['director'].apply(lambda x:x.split(","))
movies_df['listed_in'] = movies_df['listed_in'].apply(lambda x:x.split(","))

tv_show_df['description'] = tv_show_df['description'].apply(lambda x:x.split())
tv_show_df['rating'] = tv_show_df['rating'].apply(lambda x:x.split())
tv_show_df['country'] = tv_show_df['country'].apply(lambda x:x.split(","))
tv_show_df['director'] = tv_show_df['director'].apply(lambda x:x.split(","))
tv_show_df['listed_in'] = tv_show_df['listed_in'].apply(lambda x:x.split(","))
```

```python
movies_df['features'] = movies_df['cast']+movies_df['description']+movies_df['rating']+movies_df['country']+movies_df['director']+movies_df['listed_in']
tv_show_df['features'] = tv_show_df['cast']+tv_show_df['description']+tv_show_df['rating']+tv_show_df['country']+tv_show_df['director']+tv_show_df['listed_in']
```

```python
feature_df_movies = movies_df[['title', 'features']]
feature_df_tv_show = tv_show_df[['title', 'features']]
```

```python
feature_df_movies['features']= feature_df_movies['features'].apply(lambda x : list(filter(None, x)))
feature_df_tv_show['features']= feature_df_tv_show['features'].apply(lambda x : list(filter(None, x)))
```

```python
feature_df_movies['features']=feature_df_movies['features'].apply(lambda x : [i.replace(" ", "") for i in x])
feature_df_tv_show['features']=feature_df_tv_show['features'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
feature_df_movies['features']=feature_df_movies['features'].apply(lambda x : " ".join(x))
feature_df_tv_show['features']=feature_df_tv_show['features'].apply(lambda x : " ".join(x))
```

```python
feature_df_movies['features']=feature_df_movies['features'].apply(lambda x : x.lower())
feature_df_tv_show['features']=feature_df_tv_show['features'].apply(lambda x : x.lower())
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
vector = cv.fit_transform(feature_df_tv_show['features']).toarray()
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
import re 
import urllib.request
from urllib.request import urlopen
import unicodedata
```

```python
def remove_control_characters(s):
    return "".join(ch for ch in s if unicodedata.category(ch)[0]!="C")
```

```python
def get_trailer(title, movie_type, release_year):
    title = title.encode('ascii', 'ignore').decode('ascii')
    movie_type = movie_type.encode('ascii', 'ignore').decode('ascii')
    html = urlopen('https://www.youtube.com/results?search_query={title}+{movie_type}+{release_year}'.format(title=remove_control_characters(re.sub("[ ]", "+", title)), movie_type=re.sub("[ ]", "+", movie_type), release_year=release_year))
    print(html)
    video_ids = re.findall(r"watch\?v=(\S{11})", html.read().decode())
    if len(video_ids) > 0:
        return title, ("https://www.youtube.com/watch?v={}".format(video_ids[0]))
    else:
        return title, 'No link'
```

```python
all_trailer_link = [get_trailer(data.iloc[i]['title'], data.iloc[i]['type'],
                        data.iloc[i]['release_year']) for i in range(len(data)-1)]
```

```python
trailer_link_dict = {}
for i in all_trailer_link:
    trailer_link_dict[i[0]] = i[1]
```

```python
data['trailer_link'] = data['title'].map(trailer_link_dict)
```
